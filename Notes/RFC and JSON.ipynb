JSON package/notation

-JavaScript Object Notation
-Format to represent structured information
-They contain structured information in "Attribute-value" pairs which directly corresponds to a map or struct in go.
-The basic value types are:
Bool, number, string, array, "object"

```go
//Go struct
p1 := Person(name: "joe", addr:"a st.",
             phone: "123")
```

```go
//Equivalent JSON object
{"name": "joe", "addr":"a st.",
             "phone": "123"}
```

JSON Marshalling

Generating JSON representation from an object.

```go
type struct Person {
    name string
    addr string
    phone string
}

p1 := Person(name: "joe", addr:"a st.",
             phone: "123")

b_arr, err : = 
json.Marshal(p1)
```

Marshal() returns JSON representation of the object as []byte.
[]byte is a byte array which is mutable and is in UTF-8 format.

JSON Unmarshalling

```go
var p2 Person

err := json.Unmarshal(barr, &p2)
```

barr is a byte array that already contains the JSON object. Unmarshal function takes it as well as a go object as argument and returns "nil" if no errors occur. --> Object must "fit" JSON []byte. i.e attribute names should match and the size too.
