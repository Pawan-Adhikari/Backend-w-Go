Slice:

-Subset of an underlying array.

-Variable size, upto whole array

-Pointer indicates the start of the slice

-Length is the bumber of elements in the slice

-Capacirty is the maximum number of elements. : From start of slice to end of the array.

```go
arr := [...]string{"a","b","c","d","e","f","g"}
s1 := arr[1:3]
s2 := arr[2:5]
```

```go
len() : returns the length.

cap() : returns the capacity
```

```go
a1 := [3]string("a","b","c")
sli1 := a1[0:1]
fmt.Print(len(sli1), cap(sli1))
```

then, result : 1 3
since only one element, i.e 0th index element of string is chosen. And capacity is still 3 sinceit is defined from starting index of array to final index.

```go
	a1 := [3]string{"a", "b", "c"}
	sli1 := a1[1:2]
	fmt.Print(sli1[0],len(sli1), cap(sli1))
```

THIS WILL RETURN : 1 2


Slice is an mutable reference to the elements of the array. Changes made in the elements of the slice transfers to the corresponding element in the underlying array as well as other slices that reference to same value.

```go
sli := []int{1,2,3}
```

Since we aren't passing even ... to the declaration, the compiler understands that its a slice.

Make
• Create a slice (and array) using make ()
• 2-argument version:
specify type and length/capacity
• Init. to zero, length = capacity
sli = make ([]int, 10)
• 3-argument version: specify length and capacity separately
sli = make ( [lint, 10, 15)

Maps

-Go's implementation of hashtables



-Here, idMap is the map vairable name and the composite "map[string]int" is the composite map data type


```go
fmt.Println(idMap["joe"])// prints value corresponding to "joe" key
idMap["jane"] = 456//overwrites the original value
delete(idMap, "joe")//eliminates key value pair entirely
```

```go
id, p := idMap["joe"]
```

Two value assignment to value tests for the existence of key.
here, id is the value and p is the presence of key

```go
for key, val := range idMap{
    fmt.Println(key, val)
}
```

Range function is overloaded here to return key and value of a map.
